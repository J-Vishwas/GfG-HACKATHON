```python
vaccine=vaccine.drop(['marital_status',
       'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa',
       'household_adults', 'household_children', 'employment_industry',
       'employment_occupation','health_insurance','sex'],axis=1)
vaccine_test=vaccine_test.drop(['marital_status',
       'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa',
       'household_adults', 'household_children', 'employment_industry',
       'employment_occupation','health_insurance','sex'],axis=1)
```

```python
vaccine['age_group'].replace({'18 - 34 Years': 1, '35 - 44 Years': 2, '45 - 54 Years': 3, '55 - 64 Years': 4, '65+ Years': 5}, inplace=True)
vaccine_test['age_group'].replace({'18 - 34 Years': 1, '35 - 44 Years': 2, '45 - 54 Years': 3, '55 - 64 Years': 4, '65+ Years': 5}, inplace=True)
```

```python
vaccine.age_group.astype(float)
vaccine_test.age_group.astype(float)
```

```python
vaccine = pd.get_dummies(vaccine, drop_first=True)
vaccine_test = pd.get_dummies(vaccine, drop_first=True)
```

```python
vaccine=vaccine.dropna(axis=0)
vaccine_test=vaccine_test.dropna(axis=0)
```

```python

vaccine_test.shape
```

```python
from sklearn.model_selection import train_test_split
labels_path = '/training_set_labels.csv'
labels = pd.read_csv(labels_path)

# Merge features and labels
data = pd.merge(vaccine, labels, on='respondent_id')

# Define features and target variables
X = data.drop([ 'xyz_vaccine', 'seasonal_vaccine', 'respondent_id'], axis=1)
y = data[['xyz_vaccine', 'seasonal_vaccine']]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.multioutput import MultiOutputClassifier

# Initialize the Naive Bayes classifier
nb = GaussianNB()
multi_nb = MultiOutputClassifier(nb, n_jobs=-1)

# Train the model
multi_nb.fit(X_train, y_train)

# Predict probabilities
y_pred_prob = multi_nb.predict_proba(X_test)

# Extract probabilities
xyz_vaccine_prob = y_pred_prob[0][:, 1]
seasonal_vaccine_prob = y_pred_prob[1][:, 1]

# Combine the probabilities into a DataFrame
y_pred_prob_combined = pd.DataFrame({
    'xyz_vaccine': xyz_vaccine_prob,
    'seasonal_vaccine': seasonal_vaccine_prob
})
```

```python
from sklearn.metrics import roc_auc_score

# Calculate ROC AUC score
roc_auc_xyz = roc_auc_score(y_test['xyz_vaccine'], xyz_vaccine_prob)
roc_auc_seasonal = roc_auc_score(y_test['seasonal_vaccine'], seasonal_vaccine_prob)
mean_roc_auc = (roc_auc_xyz + roc_auc_seasonal) / 2

print(f"ROC AUC for xyz_vaccine: {roc_auc_xyz}")
print(f"ROC AUC for seasonal_vaccine: {roc_auc_seasonal}")
print(f"Mean ROC AUC: {mean_roc_auc}")

```
